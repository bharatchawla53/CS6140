## 1. Anomaly Detection (30 points)

**Part A (5 Points):**

By dividing a data set into quartiles, IQR is used to measure variability. The data is sorted ascending and divided into four equal parts. Q1, Q2, Q3, also known as the first, second, and third quartiles, are the values that separate the four equal parts.

Use the following data points to calculate outliers in the data
data = [11, 3, 8, 10, 12, 5, 1, 50]

Using a box plot, show the outliers in the box plot.

````python
data = [11, 3, 8, 10, 12, 5, 1, 50]

# sorted array
sorted_data = sorted(data)

# calculate q1, q2, q3 and IQR 
q1 = np.percentile(sorted_data, 25)
q2 = np.percentile(sorted_data, 50)
q3 = np.percentile(sorted_data, 75)
IQR = q3 - q1

print("Interquartile Range: ", IQR)

# calculate outliers which is defined by a range: Lower Bound: (Q1 - 1.5 * IQR) and Upper Bound: (Q3 + 1.5 * IQR)
lower_bound = (q1 - 1.5 * IQR)
upper_bound = (q3 + 1.5 * IQR)

# identify outliers
outliers = [x for x in sorted_data if x < lower_bound or x > upper_bound]
print("Outliers: ", outliers)

# create a box plot to show the outliers 
plt.figure(figsize=(6,4))
sns.boxplot(sorted_data)
plt.show()
````

####  Identify the outliers using Z-score for SalePrice with 4 different thresholds

````python
# calculate mean 
mean = np.mean(df['SalePrice'])

# calculate std
std = np.std(df['SalePrice'])

# calculate z-scores
z_scores = [(x - mean) / std for x in df['SalePrice']]

# create a new df for z-scores
z_df = pd.DataFrame(z_scores, columns=['Z-score'])

# concat original df and z_df 
df_plus_z = pd.concat([df, z_df], axis = 1)
````

````python
# returns grubbs threshold 
def grubbs_threshold(alpha):
    t_val = t.ppf(1 - alpha / (2 * N), N - 2)
    grubbs_threshold = ((N-1)/np.sqrt(N)) * (np.sqrt(t_val**2 / ((N-2 + t_val**2))))
    return grubbs_threshold
````

````python
alphas = [0.05, 0.01, 0.001, 0.0001]

# enumerate through alphas to get four different grubbs thresholds
for index, alpha in enumerate(alphas):
    column_name = "Threshold_" + str(index + 1)
    threshold = grubbs_threshold(alpha)
    df_plus_z[column_name] = df_plus_z['Z-score'].abs() >= threshold
    
df_plus_z
````

````python
# visualize outliers
plt.figure(figsize=(10, 3))
fig, ax = plt.subplots(8, sharex=True)
fig.set_figheight(35)
fig.set_figwidth(20)
colors = {True:'orangered', False:'darkblue'}

ax[0].scatter(x=range(N), y=df_plus_z['SalePrice'], c=df_plus_z['Threshold_1'].map(colors))
ax[0].set_ylabel('SalePrice')
ax[1].scatter(x=range(N), y=df_plus_z['Z-score'], c=df_plus_z['Threshold_1'].map(colors))
ax[1].set_ylabel('Z score')

ax[2].scatter(x=range(N), y=df_plus_z['SalePrice'], c=df_plus_z['Threshold_2'].map(colors))
ax[2].set_ylabel('SalePrice')
ax[3].scatter(x=range(N), y=df_plus_z['Z-score'], c=df_plus_z['Threshold_2'].map(colors))
ax[3].set_ylabel('Z score')

ax[4].scatter(x=range(N), y=df_plus_z['SalePrice'], c=df_plus_z['Threshold_3'].map(colors))
ax[4].set_ylabel('SalePrice')
ax[5].scatter(x=range(N), y=df_plus_z['Z-score'], c=df_plus_z['Threshold_3'].map(colors))
ax[6].set_ylabel('Z score')

ax[6].scatter(x=range(N), y=df_plus_z['SalePrice'], c=df_plus_z['Threshold_4'].map(colors))
ax[6].set_ylabel('SalePrice')
ax[7].scatter(x=range(N), y=df_plus_z['Z-score'], c=df_plus_z['Threshold_4'].map(colors))
ax[7].set_ylabel('Z score')
````

  2. Do Eigen decomposition using any LA library of your choice. Display scree plot. (10 points)

````python
# get the square matrix
cov_matrix = np.cov(X_standardized.T)

# calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# sort eigenvalues in descending order and corresponding vectors
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# explained variance ratio for each component
explained_variance = eigenvalues / eigenvalues.sum()
````

  3. Primary Component Selection. (Select the first 6 components) (5 points)

````python
# select first 6 components
n_components = 6
selected_eigenvectors = eigenvectors[:, :n_components]
selected_eigenvalues = eigenvalues[:n_components]

# explained variance ratio for each component
explained_variance = selected_eigenvalues / selected_eigenvalues.sum()

# scree plot
plt.figure(figsize=(10, 6))
plt.plot(np.arange(1, n_components + 1), explained_variance, marker='o', linestyle='-')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained (%)')
plt.show()
````

  4. Projection in a New Feature Space. (5 points)

````python
# project standardized data onto the new feature space defined by the primary component selection
projected_data = np.dot(X_standardized, selected_eigenvectors)

print(projected_data)
````

````python
random_seed=36784765
np.random.seed(random_seed)

# parameters for two gaussian distribution
Mean1 = 9.0  
Standard_dev1 = 5.0 
Mean2 = 2.0
Standard_dev2 = 2.0

# generate data
y1 = np.random.normal(Mean1, Standard_dev1, 500)
y2 = np.random.normal(Mean2, Standard_dev2, 2000)
data=np.append(y1,y2)
````

(1) Using a single Gaussion to estimate and draw a picure to see the result: (5 points)

````python
class Gaussian:
    
    def __init__(self, mu, sigma):
        self.mu = mu
        self.sigma = sigma
    
    def pdf(self, datum):
        exponent = np.exp(-((datum - self.mu)**2 / (2 * self.sigma**2 )))
        return (1 / (np.sqrt(2 * np.pi) * self.sigma)) * exponent
````

````python
# initialize Gaussian class
gaussian1 = Gaussian(Mean1, Standard_dev1)
gaussian2 = Gaussian(Mean2, Standard_dev2)

# get probability distribution values
pdf1 = gaussian1.pdf(data)
pdf2 = gaussian2.pdf(data)

# generate values along the x-axis
x_values = np.linspace(min(data), max(data), 1000)

# calculate the corresponding probability density values for each Gaussian
pdf_values1 = gaussian1.pdf(x_values)
pdf_values2 = gaussian2.pdf(x_values)

# plot the Gaussian distributions
plt.plot(x_values, pdf_values1, linestyle = 'dashed', color='blue', label='Gaussian 1 (Mean={}, Std={})'.format(Mean1, Standard_dev1))
plt.plot(x_values, pdf_values2, linestyle = 'dashed', color='green', label='Gaussian 2 (Mean={}, Std={})'.format(Mean2, Standard_dev2))

# plot histogram of the data for comparison
plt.hist(y1, bins=50, density=True, alpha=0.5, color='blue', label='Histogram of Gaussian 1')
plt.hist(y2, bins=50, density=True, alpha=0.5, color='green', label='Histogram of Gaussian 2')

# add labels and title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Gaussian Distributions and Histogram of Data')
plt.legend()
plt.show()
````

(2) Using a 2 Gaussian mixture model to estimate and draw a picture to see the result(Do not use sklearn GaussianMixture): (30 points)

````python
class GaussianMixture_self:
    "Model mixture of two univariate Gaussians and their EM estimation"

    def __init__(self, data, mu_min=min(data), mu_max=max(data), sigma_min=1, sigma_max=1, mix=.5):
        self.data = data
        self.mu = [mu_min, mu_max]
        self.sigma = [sigma_min, sigma_max]
        self.mix = mix
            
    def Estep(self):
        weights = np.zeros((len(self.data), 2))
        
        for i in range(2):
            weights[:, i] = self.mix * self.pdf(self.data, self.mu[i], self.sigma[i])

        return weights / np.sum(weights, axis=1, keepdims=True)
    
    def Mstep(self, weights):
        for i in range(2):
            resp_sum = np.sum(weights[:, i])
            self.mu[i] = np.sum(weights[:, i] * self.data) / resp_sum
            self.sigma[i] = np.sqrt(np.sum(weights[:, i] * (self.data - self.mu[i]) ** 2) / resp_sum)
            self.mix = resp_sum / len(self.data)

    def iterate(self, N=1, verbose=False):
        for i in range(N):
            weights = self.Estep()
            self.Mstep(weights)
            
            if verbose and i % 10 == 0: 
                model.plot()
                print("Mean: ", self.mu)
                print("Variance: ", self.sigma)
                print("Mixing Coefficients: ", self.mix)
                print("")

    def pdf(self, x, mu, sigma):
        exponent = np.exp(-((x - mu)**2 / (2 * sigma**2 )))
        return (1 / (np.sqrt(2 * np.pi) * sigma)) * exponent
    
    def plot(self):        
        # plot histogram of the data for comparison
        plt.hist(y1, bins=50, density=True, alpha=0.5, color='blue', label='Histogram of Gaussian 1')
        plt.hist(y2, bins=50, density=True, alpha=0.5, color='green', label='Histogram of Gaussian 2')

        # Plot estimated mixture model
        x = np.linspace(min(self.data), max(self.data), 1000)
        mixture_pdf = self.mix * self.pdf(x, self.mu[0], self.sigma[0]) + \
                      self.mix * self.pdf(x, self.mu[1], self.sigma[1])
        plt.plot(x, mixture_pdf, 'r--', label='Estimated Mixture Model')

        plt.title('Histogram of Data and Estimated Mixture Model')
        plt.xlabel('Data')
        plt.ylabel('Probability Density')
        plt.legend()
        plt.grid(True)
        plt.show()
````

````python
# initialize Gaussian Mixture Model
model = GaussianMixture_self(data)

# perform EM iterations
model.iterate(N=50, verbose = True)

# plot the result
model.plot()
````
